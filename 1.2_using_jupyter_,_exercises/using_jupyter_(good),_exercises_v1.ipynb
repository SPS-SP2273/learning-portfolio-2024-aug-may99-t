```python
%timeit squares_1 = [i**2 for i in range(10)]
#print(squares_1)
```

```python
%%timeit
squares_2 = list()
for i in range(10):
   squares_2.append(i**2)
#print(squares_2)
```

squares_1 may take less time to run than squares_2 because all the squares were added into the list at once. In squares 2, the .append function is used which appends each square one by one in the for loop.
